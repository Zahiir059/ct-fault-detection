```python
# DATA PREPROCESSING & NORMALIZATION
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
import joblib

# 1. Load Dataset
df = pd.read_csv("ct_fault_dataset.csv")

# 2. Dataset Information
num_samples = df.shape[0]
num_features = df.shape[1] - 1
feature_names = df.columns[:-1].tolist()
class_names = df['fault_type'].unique().tolist()
num_classes = len(class_names)

print("Number of Samples:", num_samples)
print("Number of Features:", num_features)
print("Feature Names:", feature_names)
print("Number of Classes:", num_classes)
print("Class Names:", class_names)

# 3. Split Features & Labels
X = df.drop('fault_type', axis=1)
y = df['fault_type']

# 4. Train-Test Split (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42
)

print("\nTrain Shape:", X_train.shape)
print("Test Shape:", X_test.shape)

# 5. Normalization (Standardization)
scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.transform(X_test)

# Convert normalized arrays to DataFrames
X_train_norm = pd.DataFrame(X_train_norm, columns=feature_names)
X_test_norm = pd.DataFrame(X_test_norm, columns=feature_names)

# 6. Add Output Labels Back (for inspection)
train_df = X_train_norm.copy()
train_df["fault_type"] = y_train.values

test_df = X_test_norm.copy()
test_df["fault_type"] = y_test.values

print("\nFinal TRAIN Dataset Sample:")
print(train_df.head())

print("\nFinal TEST Dataset Sample:")
print(test_df.head())

# 7. Label Encoding for Target Variable (fault_type)
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# 8. Train Random Forest Model
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train_norm, y_train_encoded)

# 9. Save the model and label encoder
joblib.dump(model, 'ct_fault_model.pkl')
joblib.dump(scaler, 'scaler.pkl')  # Save the scaler for normalization in the Streamlit app
joblib.dump(label_encoder, 'label_encoder.pkl')  # Save the label encoder for predictions

print("Model and LabelEncoder saved successfully.")
```

```python
# -----------------------
#  REGULARIZED DECISION TREE
# -----------------------

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Regularized Decision Tree (controls overfitting)
dt_model = DecisionTreeClassifier(
    max_depth=6,          # limit depth
    min_samples_leaf=10,  # minimum samples per leaf
    ccp_alpha=0.01,       # pruning strength
    random_state=42
)

dt_model.fit(X_train_norm, y_train)

# Predictions
y_train_pred_dt = dt_model.predict(X_train_norm)
y_test_pred_dt = dt_model.predict(X_test_norm)

print("Decision Tree (Regularized) - Training Accuracy:", accuracy_score(y_train, y_train_pred_dt))
print("Decision Tree (Regularized) - Testing Accuracy :", accuracy_score(y_test, y_test_pred_dt))

# Confusion Matrix
cm_dt = confusion_matrix(y_test, y_test_pred_dt, labels=dt_model.classes_)
print("\nDecision Tree (Regularized) - Confusion Matrix:\n", cm_dt)

plt.figure(figsize=(7,5))
sns.heatmap(cm_dt, annot=True, fmt="d",
            xticklabels=dt_model.classes_,
            yticklabels=dt_model.classes_,
            cmap="Greys")  # GREY
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Regularized Decision Tree - Confusion Matrix")
plt.show()

print("\nDecision Tree (Regularized) - Classification Report:")
print(classification_report(y_test, y_test_pred_dt))
```

```python
# REGULARIZED RANDOM FOREST (ENSEMBLE)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Regularized Random Forest (controls overfitting)
rf_model = RandomForestClassifier(
    n_estimators=80,      # number of trees
    max_depth=8,          # limit depth
    min_samples_leaf=5,   # minimum samples per leaf
    max_features='sqrt',  # subset of features at each split
    random_state=42
)

rf_model.fit(X_train_norm, y_train)

# Predictions
y_train_pred_rf = rf_model.predict(X_train_norm)
y_test_pred_rf = rf_model.predict(X_test_norm)

print("Random Forest (Regularized) - Training Accuracy:", accuracy_score(y_train, y_train_pred_rf))
print("Random Forest (Regularized) - Testing Accuracy :", accuracy_score(y_test, y_test_pred_rf))

# Confusion Matrix
cm_rf = confusion_matrix(y_test, y_test_pred_rf, labels=rf_model.classes_)
print("\nRandom Forest (Regularized) - Confusion Matrix:\n", cm_rf)

plt.figure(figsize=(7,5))
sns.heatmap(cm_rf, annot=True, fmt="d",
            xticklabels=rf_model.classes_,
            yticklabels=rf_model.classes_,
            cmap="Reds")  # RED
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Regularized Random Forest - Confusion Matrix")
plt.show()

print("\nRandom Forest (Regularized) - Classification Report:")
print(classification_report(y_test, y_test_pred_rf))
```

```python

# BOX 4: LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

log_model = LogisticRegression(
    max_iter=500,
    multi_class='multinomial',
    C=1.0  # L2 regularization strength (you can lower C for stronger regularization)
)
log_model.fit(X_train_norm, y_train)

y_train_pred_log = log_model.predict(X_train_norm)
y_test_pred_log = log_model.predict(X_test_norm)

print("Logistic Regression - Training Accuracy:", accuracy_score(y_train, y_train_pred_log))
print("Logistic Regression - Testing Accuracy :", accuracy_score(y_test, y_test_pred_log))

cm_log = confusion_matrix(y_test, y_test_pred_log, labels=log_model.classes_)
print("\nLogistic Regression - Confusion Matrix:\n", cm_log)

plt.figure(figsize=(7,5))
sns.heatmap(cm_log, annot=True, fmt="d",
            xticklabels=log_model.classes_,
            yticklabels=log_model.classes_,
            cmap="Blues")  # BLUE
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Logistic Regression - Confusion Matrix")
plt.show()

print("\nLogistic Regression - Classification Report:")
print(classification_report(y_test, y_test_pred_log))
```

```python
# -----------------------
# KNN CLASSIFIER

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

knn_model = KNeighborsClassifier(n_neighbors=5)  # higher k = smoother, less overfitting
knn_model.fit(X_train_norm, y_train)

y_train_pred_knn = knn_model.predict(X_train_norm)
y_test_pred_knn = knn_model.predict(X_test_norm)

print("KNN - Training Accuracy:", accuracy_score(y_train, y_train_pred_knn))
print("KNN - Testing Accuracy :", accuracy_score(y_test, y_test_pred_knn))

cm_knn = confusion_matrix(y_test, y_test_pred_knn, labels=knn_model.classes_)
print("\nKNN - Confusion Matrix:\n", cm_knn)

plt.figure(figsize=(7,5))
sns.heatmap(cm_knn, annot=True, fmt="d",
            xticklabels=knn_model.classes_,
            yticklabels=knn_model.classes_,
            cmap="Greens")  # GREEN
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("KNN - Confusion Matrix")
plt.show()

print("\nKNN - Classification Report:")
print(classification_report(y_test, y_test_pred_knn))
```

```python
# -----------------------
# LINEAR REGRESSION (ACCURACY + METRICS)
# -----------------------

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# 1. Encode target labels
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

# 2. Train Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train_norm, y_train_enc)

# 3. Predict continuous values
y_train_pred_lin = lin_reg.predict(X_train_norm)
y_test_pred_lin = lin_reg.predict(X_test_norm)

# 4. Convert continuous predictions  nearest integer class
y_train_pred_class = np.clip(np.round(y_train_pred_lin).astype(int), 0, len(le.classes_)-1)
y_test_pred_class  = np.clip(np.round(y_test_pred_lin).astype(int), 0, len(le.classes_)-1)

# 5. Decode numeric predictions  original labels
y_train_pred_final = le.inverse_transform(y_train_pred_class)
y_test_pred_final  = le.inverse_transform(y_test_pred_class)

# 6. Accuracy
train_acc = accuracy_score(y_train, y_train_pred_final)
test_acc  = accuracy_score(y_test, y_test_pred_final)

# 7. Regression Metrics
train_mse = mean_squared_error(y_train_enc, y_train_pred_lin)
test_mse  = mean_squared_error(y_test_enc,  y_test_pred_lin)

train_mae = mean_absolute_error(y_train_enc, y_train_pred_lin)
test_mae  = mean_absolute_error(y_test_enc,  y_test_pred_lin)

train_rmse = np.sqrt(train_mse)
test_rmse  = np.sqrt(test_mse)

train_r2 = r2_score(y_train_enc, y_train_pred_lin)
test_r2  = r2_score(y_test_enc,  y_test_pred_lin)

# -----------------------
# PRINT RESULTS
# -----------------------

print("Linear Regression (Classification + Regression Metrics)\n")

print("Training Accuracy :", train_acc)
print("Testing Accuracy  :", test_acc)

print("\n--- Regression Metrics ---")
print("Train MSE :", train_mse)
print("Test MSE  :", test_mse)

print("Train MAE :", train_mae)
print("Test MAE  :", test_mae)

print("Train RMSE:", train_rmse)
print("Test RMSE :", test_rmse)

print("Train R  :", train_r2)
print("Test R   :", test_r2)
```

```python
# -----------------------
#  SVM CLASSIFIER
# -----------------------

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

svm_model = SVC(
    kernel='rbf',
    C=1.0,        # lower C => stronger regularization
    gamma='scale',
    random_state=42
)
svm_model.fit(X_train_norm, y_train)

y_train_pred_svm = svm_model.predict(X_train_norm)
y_test_pred_svm = svm_model.predict(X_test_norm)

print("SVM - Training Accuracy:", accuracy_score(y_train, y_train_pred_svm))
print("SVM - Testing Accuracy :", accuracy_score(y_test, y_test_pred_svm))

cm_svm = confusion_matrix(y_test, y_test_pred_svm, labels=svm_model.classes_)
print("\nSVM - Confusion Matrix:\n", cm_svm)

plt.figure(figsize=(7,5))
sns.heatmap(cm_svm, annot=True, fmt="d",
            xticklabels=svm_model.classes_,
            yticklabels=svm_model.classes_,
            cmap="Oranges")  # ORANGE
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("SVM - Confusion Matrix")
plt.show()

print("\nSVM - Classification Report:")
print(classification_report(y_test, y_test_pred_svm))
```
