```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 1. Load Dataset
df = pd.read_csv("ct_fault_dataset.csv")

# 2. Dataset Information
num_samples = df.shape[0]
num_features = df.shape[1] - 1
feature_names = df.columns[:-1].tolist()
class_names = df['fault_type'].unique().tolist()
num_classes = len(class_names)

print("Number of Samples:", num_samples)
print("Number of Features:", num_features)
print("Feature Names:", feature_names)
print("Number of Classes:", num_classes)
print("Class Names:", class_names)

# 3. Split Features & Labels
X = df.drop('fault_type', axis=1)
y = df['fault_type']

# 4. Train-Test Split (80/20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)
print("\nTrain Shape:", X_train.shape)
print("Test Shape:", X_test.shape)

# 5. Normalization (Standardization)
scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.transform(X_test)

# Convert normalized arrays to DataFrames
X_train_norm = pd.DataFrame(X_train_norm, columns=feature_names)
X_test_norm = pd.DataFrame(X_test_norm, columns=feature_names)

# 6. Add Output Labels Back (for inspection)
train_df = X_train_norm.copy()
train_df["fault_type"] = y_train.values

test_df = X_test_norm.copy()
test_df["fault_type"] = y_test.values

print("\nFinal TRAIN Dataset Sample:")
print(train_df.head())

print("\nFinal TEST Dataset Sample:")
print(test_df.head())

# 7. Label Encoding for Target Variable (fault_type)
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# 8. Train Random Forest Model
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train_norm, y_train_encoded)

# 9. Save the model, scaler, and label encoder
joblib.dump(model, 'ct_fault_model.pkl')
joblib.dump(scaler, 'scaler.pkl')  # Save the scaler for normalization
joblib.dump(label_encoder, 'label_encoder.pkl')  # Save the label encoder

print("Model and LabelEncoder saved successfully.")

# -----------------------  MODEL TRAINING AND EVALUATION -----------------------

# Function to calculate accuracy and confusion matrix
def evaluate_model(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Accuracy
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    print(f"\nAccuracy on Training Set: {train_accuracy*100:.2f}%")
    print(f"Accuracy on Test Set: {test_accuracy*100:.2f}%")

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_test_pred)
    print(f"\nConfusion Matrix:\n{cm}")

    # Plot confusion matrix
    plt.figure(figsize=(7, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"{model.__class__.__name__} - Confusion Matrix")
    plt.show()

    # Classification Report
    print(f"\nClassification Report:\n{classification_report(y_test, y_test_pred, target_names=class_names)}")

# 8. Train Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
rf_model.fit(X_train_norm, y_train_encoded)

# Evaluate Random Forest
evaluate_model(rf_model, X_train_norm, y_train_encoded, X_test_norm, y_test_encoded)

# 9. Train Decision Tree Model
dt_model = DecisionTreeClassifier(max_depth=6, min_samples_leaf=10, ccp_alpha=0.01, random_state=42)
dt_model.fit(X_train_norm, y_train_encoded)

# Evaluate Decision Tree
evaluate_model(dt_model, X_train_norm, y_train_encoded, X_test_norm, y_test_encoded)

# 10. Train Logistic Regression Model
log_model = LogisticRegression(max_iter=500, multi_class='multinomial', C=1.0)
log_model.fit(X_train_norm, y_train_encoded)

# Evaluate Logistic Regression
evaluate_model(log_model, X_train_norm, y_train_encoded, X_test_norm, y_test_encoded)

# 11. Train KNN Model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_norm, y_train_encoded)

# Evaluate KNN
evaluate_model(knn_model, X_train_norm, y_train_encoded, X_test_norm, y_test_encoded)

# 12. Train SVM Model
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(X_train_norm, y_train_encoded)

# Evaluate SVM
evaluate_model(svm_model, X_train_norm, y_train_encoded, X_test_norm, y_test_encoded)

# 13. Train Naive Bayes Model
nb_model = GaussianNB()
nb_model.fit(X_train_norm, y_train_encoded)

# Evaluate Naive Bayes
evaluate_model(nb_model, X_train_norm, y_train_encoded, X_test_norm, y_test_encoded)

# -----------------------  USER INPUT FOR TESTING RANDOM FOREST -----------------------
print("\nPlease enter values for the features:")

user_input = {}
for feature in feature_names:
    user_input[feature] = float(input(f"Enter value for {feature}: "))  # Collecting user input

user_sample = pd.DataFrame([user_input])

# Normalize user input (same as model preprocessing)
user_input_normalized = scaler.transform(user_sample)

# Predict with the trained Random Forest model
predicted_class_idx = rf_model.predict(user_input_normalized)[0]
predicted_class = label_encoder.inverse_transform([predicted_class_idx])[0]

print(f"\nPredicted Fault Type for the entered sample using RandomForest: {predicted_class}")

# -----------------------  SAVE MODELS -----------------------
# Save the model, scaler, and label encoder
joblib.dump(rf_model, 'ct_fault_rf_model.pkl')
joblib.dump(scaler, 'scaler.pkl')  # Save the scaler for normalization
joblib.dump(label_encoder, 'label_encoder.pkl')  # Save the label encoder
```
